To Import the dependencies and seaborn itself in ypur code use the following code

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy import stats
```

### Visualizing statistical relationships
This means we are trying to understand relationship between different variables in a dataset and how the relationship affects or depends on the other variables.

```python
# importing the the Predict the number of upvotes dataset
df = pd.read_csv('upvote_train.csv')
df.head()
```

### Scatterplot using Seaborn
This shows the relationship between two numeric variables, it shows the joint distribution of two variables using cloud points
we will make use of replot() function in seaborn library.

```python
sns.relplot(x='Views', y='Upvotes', data = df)
```

next ,if we want to see the tag associated with the data

```python
sns.relplot(x='Views', y= 'Upvotes', hue= 'Tag', data = df)
```

The Hue semantic is categorical, that is the reason for different color pallete. if it is numeric the colouring becomes sequential

```python
sns.relplot(x= 'Views', y='Upvotes', hue='Answers', data=df)
```

We can also change the size of each point

```python
sns.relplot(x='Views', y='Upvotes', size='Tag', data=df)
```

We can also change the size manually by using another parameter sizes as sizes=(15,200)

```python
sns.relplot(x='Views', y='Upvotes', sizes=(15,200), data=df)
plt.show()
```

 We will be using catplot() function of seaborn library to draw the plots of categorical data.
 
 ## Jitter Plot
 for jitter plot we'll be using another dataset from the problem.HR analysis challenge, 
 lets import the dataset now

```python
df2 = pd.read_csv('hr_train.csv')
df2.head()
```

```python
df2.dtypes
```

Let see the relationship betwen a categorical and a numerical variable.
we will be using the catplot() function

```python
sns.catplot(x= 'education', y='avg_training_score', data= df2)
```

```python
# introducing another parameter "jitter"
sns.catplot(x='education', y='avg_training_score', jitter = False, data=df2)
```

## Hue Plot
Next, if we want to introduce another variable or another dimension in our plot, we can use the hue parameter.
To see the gender distribution in our plot of education and avg_training_score.

```python
sns.catplot(x='education', y='avg_training_score', hue='gender', data=df2)
plt.show()
```

```python
sns.catplot(x= 'education', y= 'avg_training_score', kind = 'swarm', data=df2)
```
